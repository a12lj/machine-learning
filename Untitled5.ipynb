### 2.5 Operadores

**Operadores aritméticos**

| Operador | Descripción       |
|----------|-------------------|
| `a + b`  | Suma              |
| `a - b`  | Resta             |
| `a / b`  | División          |
| `a // b` | División entera   |
| `a % b`  | Módulo / Resto    |
| `a * b`  | Multiplicación    |
| `a ** b` | Exponenciación    |

````python
x = 3
y = 2
print('x + y = ', x + y)   # 5
print('x - y = ', x - y)   # 1
print('x * y = ', x * y)   # 6
print('x / y = ', x / y)   # 1.5
print('x // y = ', x // y) # 1
print('x % y = ', x % y)   # 1
print('x ** y = ', x ** y) # 9
````

**Operadores de comparación**

| Operador  | Descripción       |
|-----------|-------------------|
| `a > b`   | Mayor             |
| `a < b`   | Menor             |
| `a == b`  | Igualdad          |
| `a != b`  | Desigualdad       |
| `a >= b`  | Mayor o Igual     |
| `a <= b`  | Menor o Igual     |

````python
x = 10
y = 12
print('x > y es ', x > y)   # False
print('x < y es ', x < y)   # True
print('x == y es ', x == y) # False
print('x != y es ', x != y) # True
print('x >= y es ', x >= y) # False
print('x <= y es ', x <= y) # True
````

## 3 Tipos de datos compuestos (colecciones)

### 3.1 Listas

- Una colección de objetos.
- **Mutables.**
- Tipos arbitrarios heterogéneos.
- Puede contener duplicados.
- No tienen tamaño fijo.
- Los elementos van ordenados por posición.
- Se acceden usando la sintaxis: `[index]`.
- Los índices van de `0` a `n-1`.
- Son un tipo de **Secuencia**, al igual que los strings.
- Soportan anidamiento.
- Son una implementación del tipo abstracto de datos: **Array Dinámico**.

**Operaciones con listas**


````python
# Creación de listas
letras = ['a', 'b', 'c', 'd']
palabras = 'Hola mundo como estas'.split()
numeros = list(range(7))
print(letras)    # ['a', 'b', 'c', 'd']
print(palabras)  # ['Hola', 'mundo', 'como', 'estas']
print(numeros)   # [0, 1, 2, 3, 4, 5, 6]
print(type(numeros))  # <class 'list'>
````

````python
# Pueden contener elementos arbitrarios / heterogéneos
mezcla = [1, 3.4, 'a', None, False]
print(mezcla)      # [1, 3.4, 'a', None, False]
print(len(mezcla)) # 5
````

````python
# Pueden incluso contener objetos más "complejos"
lista_con_funcion = [1, 2, len, pow]
print(lista_con_funcion)
````

````python
# Pueden contener duplicados
lista_con_duplicados = [1, 2, 3, 3, 3, 4]
print(lista_con_duplicados)
````

````python
# Longitud de una lista
letras = ['a', 'b', 'c', 'd', 1]
print(len(letras))  # 5
````

````python
# Acceso a un elemento
print(letras[2])   # c
print(letras[-5])  # a
print(letras[0])   # a
````

````python
# Slicing: lista[inicio:fin:paso]
letras = ['a', 'b', 'c', 'd', 'e']
print(letras[1:3])   # ['b', 'c']
print(letras[:3])    # ['a', 'b', 'c']
print(letras[:-1])   # ['a', 'b', 'c', 'd']
print(letras[2:])    # ['c', 'd', 'e']
print(letras[:])     # ['a', 'b', 'c', 'd', 'e']
print(letras[::2])   # ['a', 'c', 'e']
````

````python
# Copiar una lista con slicing
letras = ['a', 'b', 'c', 'd']
a = letras[:]
print(a)              # ['a', 'b', 'c', 'd']
print(letras.copy())  # ['a', 'b', 'c', 'd']
````

````python
# Añadir un elemento al final
letras.append('e')
print(letras)

# También con +=
letras += 'e'
print(letras)
````

````python
# Insertar en posición
letras.insert(1, 'g')
print(letras)
````

````python
# Modificación individual
letras[5] = 'f'
print(letras)

# Modificación múltiple con slicing
letras = ['a', 'b', 'c', 'f', 'g', 'h', 'i', 'j', 'k']
letras[0:7:2] = ['z', 'x', 'y', 'p']
print(letras)  # ['z', 'b', 'x', 'f', 'y', 'h', 'p', 'j', 'k']
````

````python
# Anidamiento con asignación individual
numeros = [1, 2, 3]
numeros[1] = [10, 20, 30]
print(numeros)          # [1, [10, 20, 30], 3]
print(numeros[1][0])    # 10
numeros[1][2] = [100, 200]
print(numeros)          # [1, [10, 20, [100, 200]], 3]
print(numeros[1][2][1]) # 200
````

````python
# Eliminar un elemento
if 'p' in letras:
    letras.remove('p')
print(letras)

# pop() — elimina el último elemento y lo devuelve
elemento = letras.pop()
print(elemento)
print(letras)
````

````python
# Encontrar índice de un elemento
letras = ['a', 'b', 'c', 'c']
if 'a' in letras:
    print(letras.index('a'))  # 0
````

````python
# Concatenar listas
lacteos = ['queso', 'leche']
frutas = ['naranja', 'manzana']
compra = lacteos + frutas
print(compra)  # ['queso', 'leche', 'naranja', 'manzana']

# Sin crear nueva lista
frutas.extend(['pera', 'uvas'])
print(frutas)  # ['naranja', 'manzana', 'pera', 'uvas']

# Anidar sin crear nueva lista
frutas.append(['pera', 'uvas'])
print(frutas)  # ['naranja', 'manzana', ['pera', 'uvas']]
````

````python
# Replicar una lista
lacteos = ['queso', 'leche']
print(lacteos * 3)
# ['queso', 'leche', 'queso', 'leche', 'queso', 'leche']
````

````python
# Ordenar una lista
lista = [4, 3, 8, 1]
lista.sort()
print(lista)               # [1, 3, 4, 8]
lista.sort(reverse=True)
print(lista)               # [8, 4, 3, 1]

# sorted() no modifica la lista original
compra = ['Huevos', 'Pan', 'zapallo', 'Leche', 'Licor']
print(sorted(compra))  # ['Huevos', 'Leche', 'Licor', 'Pan', 'zapallo']
print(compra)          # ['Huevos', 'Pan', 'zapallo', 'Leche', 'Licor']
````

````python
# Pertenencia
lista = [1, 2, 3, 4]
print(1 in lista)  # True
print(5 in lista)  # False
````

````python
# Anidamiento
letras = ['a', 'b', 'c', ['x', 'y', ['i', 'j', 'k']]]
print(letras[0])       # a
print(letras[3][0])    # x
print(letras[3][2][0]) # i
````

**Alias / Referencias en las listas**

````python
lista = [2, 4, 16, 32]
ref = lista
ref[2] = 64
print(ref)    # [2, 4, 64, 32]
print(lista)  # [2, 4, 64, 32]  ← también cambia!
````

````python
# Copia superficial
lista = [2000, 4, 16, 32]
copia = lista[:]   # o lista.copy()
copia[2] = 64
print(lista)  # [2000, 4, 16, 32]
print(copia)  # [2000, 4, 64, 32]
````

````python
# Las listas anidadas se copian por referencia
lista = [2, 4, 16, 32, [34, 10, [5, 5]]]
copia = lista.copy()
copia[3] = 20
copia[4][0] = 28
print(copia)  # [2, 4, 16, 20, [28, 10, [5, 5]]]
print(lista)  # [2, 4, 16, 32, [28, 10, [5, 5]]]  ← sublista afectada!
````

````python
# Evitar que listas anidadas se copien por referencia → deepcopy
import copy
lista = [2, 4, 16, 32, [34, 10, [5, 5]]]
copia = copy.deepcopy(lista)
copia[0] = 454
copia[4][2][0] = 64
print(lista)
print(copia)
````

- Los elementos **no van ordenados** por posición.
- Solo pueden contener **objetos inmutables**.

### 3.4 Tuplas

Al igual que las listas:

- Colección de elementos, tipos arbitrarios.
- Puede contener duplicados.
- No tienen tamaño fijo.
- Los elementos van ordenados por posición; acceso con `[index]`.
- Soportan anidamiento.

A diferencia de las listas:

- Se definen por medio de **paréntesis**.
- **Inmutables.**

¿Por qué tuplas?

- Representación de una colección fija de elementos (por ejemplo, una fecha).
- Pueden usarse como claves de un diccionario.

### 3.5 Secuencias

- Formalmente, objetos iterables no materializados.
- No son listas; usar `list()` para materializarlas.
- Ejemplos: `range`, `enumerate`.

````python
# range
list(range(0, 10, 2))  # [0, 2, 4, 6, 8]

# enumerate
lista = [10, 20, 30]
print(list(enumerate(lista)))  # [(0, 10), (1, 20), (2, 30)]

for index, value in enumerate(lista):
    print('Index = ' + str(index) + '. Value = ' + str(value))
````

````python
# zip — une elemento a elemento dos colecciones
nombres = ['Manolo', 'Pepe', 'Luis']
edades = [31, 34, 34]
for nombre, edad in zip(nombres, edades):
    print('Nombre: ' + nombre + ', edad: ' + str(edad))

# unzip
jugadores = zip(nombres, edades)
ns, es = zip(*jugadores)
print(list(ns))
````

### 3.6 En Python todo son objetos

Cada objeto tiene:

- **Identidad:** nunca cambia una vez creado. Operador `is` compara identidad. Función `id()` devuelve identidad.
- **Tipo:** determina posibles valores y operaciones. Función `type()` devuelve el tipo. No cambia.
- **Valor:** pueden ser mutables o inmutables.
  - **Tipos mutables:** `list`, `dictionary`, `set` y tipos definidos por el usuario.
  - **Tipos inmutables:** `int`, `float`, `bool`, `string` y `tuple`.

````python
# Asignación
list_numbers = [1, 2, 3]    # Lista (mutable)
tuple_numbers = (1, 2, 3)   # Tupla (inmutable)
list_numbers[0] = 100       # OK
# tuple_numbers[0] = 100    # TypeError

# Identidad
list_numbers += [4, 5, 6]   # La lista original se extiende → mismo id
tuple_numbers += (4, 5, 6)  # Se crea un nuevo objeto → id diferente

# Referencias
list_numbers_2 = list_numbers  # referencia compartida
list_numbers.append(4)
print(list_numbers)    # [1, 2, 3, 4]
print(list_numbers_2)  # [1, 2, 3, 4]  ← también cambia

# Inmutables
text = "Hola"
text_2 = text
text += " Mundo"
print(text)    # Hola Mundo
print(text_2)  # Hola  ← no cambia (nuevo objeto)
````

````python
# Curiosidad: tupla inmutable con lista mutable dentro
teams = ["Team A", "Team B", "Team C"]
player = (23, teams)
teams[2] = "Team J"
print(player)  # (23, ['Team A', 'Team B', 'Team J']) ← la lista sí cambia
# player[1][0] = 'Team XX'  → esto también funciona (modifica la lista)
````
