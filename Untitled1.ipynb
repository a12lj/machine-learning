# 2. Series

- Datos unidimensionales (similar a NumPy).
- Elementos e índices modificables.

```python
countries = pd.Series(['Spain','Andorra','Gibraltar','Portugal','France']) print(countries)
```

```python
# Especificando el índice
countries = pd.Series(
    ['Spain', 'Andorra', 'Gibraltar', 'Portugal', 'France'],
    index=range(10, 60, 10)
)

print(countries)
```

```python
## Los índices pueden ser de otros tipos
football_cities = pd.Series(
    ['Barcelona', 'Madrid', 'Valencia', 'Sevilla'],
    index=['a', 'b', 'c', 'd']
)

print(football_cities)
```

```python
## Atributos de una Serie
football_cities.name = 'Ciudades con dos equipos en primera'  # Nombre de la Serie
football_cities.index.name = 'Id'                             # Nombre del índice

print(football_cities)
```

```python
# Acceso similar a NumPy o listas, según posición (deprecated)
print(football_cities[2])

# Acceso a través del índice semántico
print(football_cities['c'])

# Comparación de ambos accesos
print(football_cities['c'] == football_cities[2])
```

# 3. Tratamiento similar a `ndarray`

```python
# Múltiple recolección de elementos
print(football_cities[['a', 'c']])

print(football_cities[[0, 3]])
```

```python
# Slicing
print(football_cities[:'c'])   # Incluye ambos extremos con el índice semántico

print(football_cities[:2])     # Slice por posición
```

```python
# Cast a list
lista = list(football_cities[:'c'])

print(lista)
print(type(lista))
```

```python
type(football_cities[:'c'])
```

```python
# Cast a ndarray
import numpy as np

cities = np.array(football_cities[:'c'])

print(cities)
print(type(cities))
```

```python
# Cast a dictionary
lista = dict(football_cities[:'c'])

print(lista)
```
